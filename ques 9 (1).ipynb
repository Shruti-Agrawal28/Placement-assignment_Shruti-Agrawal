```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the data from the CSV file
url = 'https://data.wa.gov/api/views/f6w7-q2d2/rows.csv?accessType=DOWNLOAD'
df = pd.read_csv(url)
```

```python
df.head()
```

### All the cars and their types that do not qualify for clean alternative fuel vehicle

```python
non_clean_cars = df[(df["Clean Alternative Fuel Vehicle (CAFV) Eligibility"] == 'Not eligible due to low battery range') |
                    (df["Clean Alternative Fuel Vehicle (CAFV) Eligibility"] == 'Eligibility unknown as battery range has not been researched')][['Make', 'Model']]
non_clean_cars
```

### All TESLA cars with the model year, and model type made in Bothell City


```python
tesla_cars_in_bothell = df[(df['Make'] == 'TESLA') & (df['City'] == 'Bothell')][['Model Year', 'Model']]
tesla_cars_in_bothell
```

### All the cars that have an electric range of more than 100 and were made after 2015


```python
electric_cars_100_miles_plus = df[(df['Electric Range'] > 100) & (df['Model Year'] > 2015)]
electric_cars_100_miles_plus
```

### Plots to show the distribution between city and electric vehicle type


```python

grouped_data = df.groupby(['City', 'Electric Vehicle Type']).size().unstack()

# Plot the distribution using a stacked bar chart
fig, ax = plt.subplots(figsize=(30, 12))  # Adjust the height here

# Set the width of each bar
bar_width = 1.5

# Create a list of x positions for the bars
x = range(len(grouped_data.index))

# Plot each electric vehicle type as a separate bar
for i, col in enumerate(grouped_data.columns):
    ax.bar(x, grouped_data[col], width=bar_width, label=col, alpha=0.8)

# Set labels and title
plt.xlabel('City', fontsize=24)
plt.ylabel('Count', fontsize=24)
plt.title('Distribution of Electric Vehicle Types by City', fontsize=30)

# Set legend
legend_title = 'Electric Vehicle Type'
ax.legend(title=legend_title, bbox_to_anchor=(1, 1))

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Adjust the plot spacing
plt.tight_layout()

# Show the plot
plt.show()
```
